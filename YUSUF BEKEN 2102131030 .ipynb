```python
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk

nltk.download('punkt')
nltk.download('stopwords')


stop_words = set(stopwords.words('turkish'))

def tokenize_ve_temizle(sarki_metni):
    sarki_metni = sarki_metni.lower()
    sarki_metni = re.sub(r"[^\w\s]", "", sarki_metni)  # noktalama temizle
    kelimeler = word_tokenize(sarki_metni)
    return [k for k in kelimeler if k not in stop_words]

tokenize_edilmis_sarkilar = [tokenize_ve_temizle(sarki) for sarki in temizlenmis_sarkilar]

```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def sarki_vektor_hesapla(tokenler):
    kelime_vektorleri = [model.wv[kelime] for kelime in tokenler if kelime in model.wv]
    if not kelime_vektorleri:
        return np.zeros(model.vector_size)
    return np.mean(kelime_vektorleri, axis=0)

```

```python
def true_false_test(sarki_1_idx, sarki_2_idx, threshold=0.7):

    v1 = sarki_vektor_hesapla(tokenize_edilmis_sarkilar[sarki_1_idx])
    v2 = sarki_vektor_hesapla(tokenize_edilmis_sarkilar[sarki_2_idx])

    benzerlik = cosine_similarity([v1], [v2])

    if benzerlik[0][0] >= threshold:
        return True  # Benzer
    else:
        return False  # Benzer değil

sarki_1_idx = 0  # 1. şarkı
sarki_2_idx = 1  # 2. şarkı

test_sonucu = true_false_test(sarki_1_idx, sarki_2_idx)
print(f"Şarkı {sarki_1_idx + 1} ve Şarkı {sarki_2_idx + 1} arasındaki benzerlik: {'True' if test_sonucu else 'False'}")
```

```python
import pandas as pd
import zeyrek
from nltk.tokenize import word_tokenize
import snowballstemmer
from nltk.corpus import stopwords


import nltk
nltk.download('stopwords')


stop_words = set(stopwords.words("turkish"))

def remove_stopwords(text):
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word not in stop_words]
    return " ".join(filtered_tokens)

analyzer = zeyrek.MorphAnalyzer()
def lemmatize_text(text):
    tokens = text.split()
    lemmatized = []
    for token in tokens:
        analysis = analyzer.analyze(token)
        if analysis:
       
            best_analysis = analysis[0] 
            lemma = best_analysis[0].lemma if len(best_analysis) > 0 else token  # Lemmaya bu şekilde erişiyoruz
            lemmatized.append(lemma)
        else:
            lemmatized.append(token) 
    return " ".join(lemmatized)


stemmer = snowballstemmer.stemmer("turkish")
def stem_text(text):
    tokens = text.split()
    stemmed = [stemmer.stemWord(token) for token in tokens]
    return " ".join(stemmed)


data = ["Bu şarkı çok güzel ve anlamlı.", "Hızlıca koşarak geldim.", "Bugün hava çok güzel!"]


clean_data = [remove_stopwords(text) for text in data]


lemmatized_data = [lemmatize_text(text) for text in clean_data]
stemmed_data = [stem_text(text) for text in clean_data]


df_lemmatized = pd.DataFrame({'text': lemmatized_data})
df_stemmed = pd.DataFrame({'text': stemmed_data})


df_lemmatized.to_csv("lemmatized_data.csv", index=False)
df_stemmed.to_csv("stemmed_data.csv", index=False)

print("Veri setleri CSV dosyasına kaydedildi.")
```

```python
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np


def zipf_analysis(text_data):
    all_text = " ".join(text_data)
    
    words = all_text.lower().split()
    
    word_counts = Counter(words)
    
    freq_sorted = sorted(word_counts.values(), reverse=True)
    
    rank = np.arange(1, len(freq_sorted) + 1)
    freq_log = np.log10(freq_sorted)
    rank_log = np.log10(rank)
    
    return rank_log, freq_log

rank_log_lemmatized, freq_log_lemmatized = zipf_analysis(lemmatized_data)
rank_log_stemmed, freq_log_stemmed = zipf_analysis(stemmed_data)

plt.figure(figsize=(10, 6))
plt.plot(rank_log_lemmatized, freq_log_lemmatized, label="Lemmatization", color='b')
plt.plot(rank_log_stemmed, freq_log_stemmed, label="Stemming", color='r')
plt.xlabel('Log(Rank)')
plt.ylabel('Log(Frequency)')
plt.title('Zipf Law: Lemmatization vs Stemming')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

def tfidf_vectorization(data):
    vectorizer = TfidfVectorizer()
    
    tfidf_matrix = vectorizer.fit_transform(data)
    
    tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())
    
    return tfidf_df

tfidf_lemmatized = tfidf_vectorization(lemmatized_data)

tfidf_stemmed = tfidf_vectorization(stemmed_data)

tfidf_lemmatized.to_csv("tfidf_lemmatized.csv", index=False)
tfidf_stemmed.to_csv("tfidf_stemmed.csv", index=False)

print("TF-IDF vektörleştirme işlemi tamamlandı ve dosyalar kaydedildi.")
```

```python
import gensim
from gensim.models import Word2Vec
import pandas as pd
from nltk.tokenize import word_tokenize

lemmatized_data = ["bu bir örnek lemmatize metindir", "diğer örnek cümleler burada"]
stemmed_data = ["bu bir örnek stemming metindir", "diğer örnek cümleler burada"]

lemmatized_tokens = [word_tokenize(text) for text in lemmatized_data]
stemmed_tokens = [word_tokenize(text) for text in stemmed_data]

parameters = [
    {'model_type': 'cbow', 'window': 2, 'vector_size': 100},
    {'model_type': 'skipgram', 'window': 2, 'vector_size': 100},
    {'model_type': 'cbow', 'window': 4, 'vector_size': 100},
    {'model_type': 'skipgram', 'window': 4, 'vector_size': 100},
    {'model_type': 'cbow', 'window': 2, 'vector_size': 300},
    {'model_type': 'skipgram', 'window': 2, 'vector_size': 300},
    {'model_type': 'cbow', 'window': 4, 'vector_size': 300},
    {'model_type': 'skipgram', 'window': 4, 'vector_size': 300}
]

def train_and_save_model(data, params, dataset_name):
    for param in parameters:
        model_type = param['model_type']
        window = param['window']
        vector_size = param['vector_size']
        
        model = Word2Vec(sentences=data, 
                         vector_size=vector_size, 
                         window=window, 
                         sg=1 if model_type == 'skipgram' else 0, 
                         min_count=1, 
                         workers=4)
        
        model_name = f"word2vec_{dataset_name}_{model_type}_win{window}_dim{vector_size}.model"
        model.save(model_name)
        print(f"Model kaydedildi: {model_name}")

train_and_save_model(lemmatized_tokens, parameters, 'lemmatized')
train_and_save_model(stemmed_tokens, parameters, 'stemmed')

print("Tüm modeller kaydedildi.")
```
